```python
import re
```

```python
def lat_long(x,index):
    lat_long_lst = re.findall("[-\d]+.[\d]+",x)
    try:
        return lat_long_lst[index]
    except:
        return ""
```

```python
df['Lat'] = [lat_long(x,0) for x in df['Lat_Long Location']]
df['Long'] = [lat_long(x,1) for x in df['Lat_Long Location']]
```
